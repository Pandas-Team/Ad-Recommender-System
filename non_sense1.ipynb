```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
```

```
def plot_3D(data, labels):
    fig = plt.figure(figsize=(8, 6))
    t = fig.suptitle('view', fontsize=14)
    ax = fig.add_subplot(111, projection='3d')

    xs = list(data['ctr'])
    ys = list(data['click'])
    zs = list(data['views'])
    data_points = [(x, y, z) for x, y, z in zip(xs, ys, zs)]
    
    unique_labels = set(labels)
    color_spect = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
    colors = [color_spect[res] for res in list(labels)]

    for data_rec, color in zip(data_points, colors):
        x, y, z = data_rec
        ax.scatter(x, y, z, alpha=0.6, c=[color], s=30)

    ax.set_xlabel('ctr')
    ax.set_ylabel('click')
    ax.set_zlabel('views')
```
